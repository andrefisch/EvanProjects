```python
from urllib.request import Request, urlopen
from datetime import *
from bs4 import BeautifulSoup as bs
import openpyxl
import pygame
import re
import time
```

```python
url = 'https://eventsinamerica.com/calendar/2018/boston-trade-shows.html'

request = Request(url)
json = urlopen(request).read()
array = str(json)
soup = bs(array)
prettyHTML = soup.prettify()
prettyHTML = prettyHTML.split('\n')
```

```python
last  = False
info  = []
event = []
count = 0
for i in range(650, len(prettyHTML)):
    short = prettyHTML[i][11:]
    #if "Source" in prettyHTML[i]:
    #    break
    if '<td>' in prettyHTML[i]:
        last = True
    else:
        if '\\r\\n' not in prettyHTML[i] and '</' not in prettyHTML[i] and '<tr>' not in prettyHTML[i]:
            print(str(i) + ": " + short)
            if short[0] == "<":
                info.append(event)
                event = []
                event.append(short[1:])
            else:
                event.append(short)
        last = False     
```

```python
EVENT      = "A"
LOCATION   = "B"
START_DATE = "C"
END_DATE   = "D"
ATTENDEES  = "E"
WEBSITE    = "F"

def create_event_spreadsheet(info):
    # Create a new excel file
    out = openpyxl.Workbook()
    # Open the worksheet we want to edit
    outsheet = out.create_sheet("EventInfo")

    # if 'sheet' appears randomly we can delete it
    rm = out.get_sheet_by_name('Sheet')
    out.remove_sheet(rm)

    #################
    # DO STUFF HERE #
    #################
    outsheet[EVENT      + '1'].value = "Event"
    outsheet[LOCATION   + '1'].value = "Location"
    outsheet[START_DATE + '1'].value = "Start Date"
    outsheet[END_DATE   + '1'].value = "End Date"
    outsheet[ATTENDEES  + '1'].value = "Attendees"
    outsheet[WEBSITE    + '1'].value = "Website"
    
    row = 2
    regexURL = "'(.*)'"
    for i in range(1, len(info)):
        event = info[i]
        event[0] = event[0].replace('amp;', '')
        event[1] = event[1].replace('amp;', '')
        matchURL = re.search(regexURL, event[0])
        if matchURL:
            website = matchURL.group(1)
        else:
            website = ""
        print(event)
        if len(event) == 5:
            outsheet[EVENT      + str(row)].value = event[1]
            outsheet[LOCATION   + str(row)].value = event[2]
            outsheet[START_DATE + str(row)].value = event[3]
            outsheet[ATTENDEES  + str(row)].value = int(event[4].replace(',', ''))
            outsheet[WEBSITE    + str(row)].value = website[8:-1]
        elif len(event) == 6:
            outsheet[EVENT      + str(row)].value = event[1]
            outsheet[LOCATION   + str(row)].value = event[2]
            outsheet[START_DATE + str(row)].value = event[3]
            outsheet[END_DATE   + str(row)].value = event[4]
            outsheet[ATTENDEES  + str(row)].value = int(event[5].replace(',', ''))
            outsheet[WEBSITE    + str(row)].value = website[8:-1]
        row = row + 1

    # Save the file
    out.save("ConventionCenterEvents.xlsx")
    
    # LMK when the script is done
    pygame.init()
    pygame.mixer.music.load('/home/andrefisch/python/evan/note.mp3')
    pygame.mixer.music.play()
    time.sleep(5)
    pygame.mixer.music.stop()

```
